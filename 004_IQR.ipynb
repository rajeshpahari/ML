# IQR: Interquartile Range

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#Create Data
n=1000
data=np.random.randn(n)**2

plt.hist(data,bins=30)
```

```python
#rankeR transform data & Sclae to 1
dataR=stats.rankdata(data)/n
plt.plot(dataR,'s')
```

```python
plt.plot(data,'p')
```

```python
plt.plot(data,dataR,'s')
```

```python


#find the indices to closest to 25%
q1=np.argmin((dataR-.25)**2)

#find the indices to closest to 75%
q3=np.argmin((dataR-.75)**2)

iqrVals=data[[q1,q3]]
iqrVals

iqrRange=iqrVals[1]-iqrVals[0]
print('Manual IQR:',iqrRange)

computedIQR=stats.iqr(data)
print('computed IQr:',computedIQR)
```
