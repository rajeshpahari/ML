## Central tendency

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# lets create some data of sample size n
n=1001
nbins=30

d1=np.random.randn(n)-1 #normal data
d2=3*np.random.randn(n)
d3=np.random.randn(n)+1 #normal data

#histogram data
y1,x1=np.histogram(d1,nbins)
x1=(x1[1:]+x1[:-1])/2

y2,x2=np.histogram(d2,nbins)
x2=(x2[1:]+x2[:-1])/2

y3,x3=np.histogram(d3,nbins)
x3=(x3[1:]+x3[:-1])/2


#Plot
plt.plot(x1,y1,'b')
plt.plot(x2,y2,'r')
plt.plot(x3,y3,'g')

plt.xlabel('x value')
plt.ylabel('data value');
```

```python
#plot mean on the graph
mean_d1=sum(d1)/len(d1)
mean_d2=np.mean(d2)
mean_d3=np.mean(d3)

#plot

plt.plot(x1,y1,'b',x2,y2,'r',x3,y3,'g')
plt.plot([mean_d1,mean_d1],[0,max(y1)],'b--')
plt.plot([mean_d2,mean_d2],[0,max(y2)],'r--')
plt.plot([mean_d3,mean_d3],[0,max(y3)],'g--')
```

Notice mean is not in peak....its not following central tendency in real

```python
# More clearly
d4=np.hstack((np.random.randn(n)-2,np.random.randn(n)+2)) #normal data


#histogram data
y4,x4=np.histogram(d4,nbins)
x4=(x4[1:]+x4[:-1])/2
mean_d4=np.mean(d4)

plt.plot(x4,y4,'g-')
plt.plot([mean_d4,mean_d4],[0,max(y1)],'b--')
```

# Does median always near to center of the data ? lets check with log normal distribution which is rightly skewed

```python
mean=.7
var=0
nbins=50
n=1000

data=mean*np.random.randn(n)+var
data=np.exp(data)

y,x=np.histogram(data,nbins)
x=(x[1:]+x[:-1])/2
mean_data=np.mean(data)
median_data=np.median(data)

plt.plot(x,y,'g-')
plt.plot([mean_data,mean_data],[0,max(y)],'b--')
plt.plot([median_data,median_data],[0,max(y)],'r--')
```

meadian here better representing central tedency than mean

# Mode Now

```python
data=np.round(np.random.randn(10))
unique_data=np.unique(data)

for i in range(len(unique_data)):
  print(f'{unique_data[i]} appears {sum(data==unique_data[i])} times')


print(' ')
print('Mode of the data is %g:'%stats.mode(data)[0][0])

```
