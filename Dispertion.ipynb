# Dispertion around central tendency: Variance ,Standard Deviation Fano Factor & Co-efficient of Variations

```python
#Import Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Lets create some ditribution

n=1001
nbins=30

d1=np.random.randn(n)-1
d2=3*np.random.randn(n)
d3=np.random.randn(n)+1


#find out histogram
y1,x1=np.histogram(d1,nbins)
x1=(x1[1:]+x1[:-1])/2

y2,x2=np.histogram(d2,nbins)
x2=(x2[1:]+x2[:-1])/2

y3,x3=np.histogram(d3,nbins)
x3=(x3[1:]+x3[:-1])/2

plt.plot(x1,y1,'r')
plt.plot(x2,y2,'b')
plt.plot(x3,y3,'g')

```

```python
mean_1=np.mean(d1)
mean_2=np.mean(d2)
mean_3=np.mean(d3)

std_1=np.std(d1,ddof=1)
std_2=np.std(d2,ddof=1)
std_3=np.std(d3,ddof=1)


plt.plot(x1,y1,'r',x2,y2,'b',x3,y3,'g')
plt.plot([mean_1,mean_1],[0,max(y1)],'r--')
plt.plot([mean_2,mean_2],[0,max(y2)],'b--')
plt.plot([mean_3,mean_3],[0,max(y3)],'g--')
plt.show()
```

```python
# Now Lets Plot Most of the dispertion of the data under 2 sigma values

mean_1=np.mean(d1)
mean_2=np.mean(d2)
mean_3=np.mean(d3)

std_1=np.std(d1,ddof=1)
std_2=np.std(d2,ddof=1)
std_3=np.std(d3,ddof=1)


plt.plot(x1,y1,'r',x2,y2,'b',x3,y3,'g')
plt.plot([mean_1,mean_1],[0,max(y1)],'r--')
plt.plot([mean_2,mean_2],[0,max(y2)],'b--')
plt.plot([mean_3,mean_3],[0,max(y3)],'g--')


plt.plot([mean_1-std_1,mean_1+std_1],[.4*max(y1),.4*max(y1)],'r',lw=10)
plt.plot([mean_2-std_2,mean_2+std_2],[.4*max(y2),.4*max(y2)],'b',lw=10)
plt.plot([mean_3-std_3,mean_3+std_3],[.4*max(y3),.4*max(y3)],'g',lw=10);
```

```python
# Different variances
var=np.arange(1,11)
N=300

varmeasures=np.zeros((4,len(var)))
varmeasures


for i in range(len(var)):

  #Create data & difference of mea from data points
  data=np.random.randn(N)*var[i]
  datadiff=data-np.mean(data)


#Variance
varmeasures[0,i]=(sum(datadiff**2))/(N-1)

#Pouplation variance or biased variance
varmeasures[1,i]=(sum(datadiff**2))/(N)

#Standard deviation
varmeasures[2,i]=np.sqrt(((sum(datadiff**2))/(N-1)))

#Mean Absolut error
varmeasures[3,i]=sum(abs(datadiff))/(N-1)


#Plot them
plt.plot(var,varmeasures.T)
plt.legend(('Var','biased var', 'stdev','MEA'))

```
