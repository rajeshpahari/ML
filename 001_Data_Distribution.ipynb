# Data Distribution

 Gaussian Distribution or Normal distribution

```python
#Libararies
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
N=1001 #sample size
x=np.linspace(-4,4,N)
gausdist=stats.norm.pdf(x)#probability density function

#Plot
plt.plot(x,gausdist)
plt.title('Analytical Gaussian/Normal Distribution')

```

```python
#is this a data or probablity distribution?? lets check sum of all probabilty
print(sum(gausdist))
```

so its not a probability distribution. Lets convert it as probability distribution

```python
#option1:
N=1001 #sample size
x=np.linspace(-4,4,N)
gausdist=stats.norm.pdf(x)#probability density function

#Plot
plt.plot(x,gausdist/sum(gausdist))
plt.title('Analytical Gaussian/Normal Distribution')
```

```python
#option 2: Scaling by dx - calculas way...hight/weight

N=1001 #sample size
x=np.linspace(-4,4,N)
gausdist=stats.norm.pdf(x)#probability density function

#Plot
plt.plot(x,(gausdist)*np.diff(x[:2]))
plt.title('Analytical Gaussian/Normal Distribution')
print(sum(gausdist)*np.diff(x[:2]))
```

#Our Own function for gaussian distribution

```python
mean=5
var=1
n=1000 #sample size

data=var*np.random.randn(n)+mean #data around mean with a variance of Var*random number

#Plot

plt.hist(data,bins=25)
plt.title('Our Own normally distributed data')
```

#                 **Unifrom data distribution**

Own function

```python
mean=.5
var=2
n=10001

data=var*np.random.rand(n)+mean-var/2 #we wil use Rand function from numpy

fig,ax=plt.subplots(1,2,figsize=(8,4))

ax[0].plot(data,'.',markersize=2)
ax[0].set_title('Uniformly distributed data value')


ax[1].hist(data,bins=25)
ax[1].set_title('Uniformly distributed data Histogram');
```

see the data is not around the mean but everywhere

# Log normal distribution

```python
# from Stats function
n=1001
x=np.linspace(0,4,n)
lognormdist=stats.lognorm.pdf(x,1)

#plot
plt.plot(x,lognormdist)
plt.title('Analytical Log normal distribution')
```

```python
#our own function
mean=5
var=.5
n=2001

data=var*np.random.randn(n)+mean
data=np.exp(data)

fig,ax=plt.subplots(1,2,figsize=(8,4))

ax[0].plot(data,'.',markersize=2)
ax[0].set_title('log normal distributed data value')


ax[1].hist(data,bins=25)
ax[1].set_title('lognormal distributed data Histogram');
```

# Binomial distribution

```python
# from Stats function
n=10
p=.5
x=range(n+2)
binomdist=stats.binom.pmf(x,n,p)

#plot
plt.bar(x,binomdist)
plt.title('Analytical binomial distribution')
```

# t test


```python
x=np.linspace(-4,4,1001)
df=200 #degree of freedom
t=stats.t.pdf(x,df)

plt.plot(x,t)
plt.title('t Distribution with degree of freedom (%g)'%df)
```

# F distribution

```python
x=np.linspace(-1,10,1001)
num_df=5
den_df=200 #degree of freedom
f=stats.f.pdf(x,num_df,den_df)

plt.plot(x,f)
plt.title('f Distribution ')
```
